```python
import pandas as pd
```

```python
df1 = pd.read_csv("/content/drive/MyDrive/metadataZip",sep="\t",header=None)
```

```python
df2 = pd.read_csv("/content/drive/MyDrive/reviewContentZip",sep="\t",header=None)
```

```python
df2.columns = ["user_id", "prod_id", "date", "review"]
```

```python
pd.merge(df1,df2,on=['user_id','prod_id','date'],how='left')
```
